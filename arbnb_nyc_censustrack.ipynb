```python
# Dependencies
import numpy as np
import pandas as pd
# DF TO EXCEL
from pandas import ExcelWriter
import math
import decimal
import requests
import time
from pprint import pprint 
from matplotlib import pyplot as plt
import os
import json
import urllib.request
import censusgeocode as cg
```

```python
lat_lng_newyork_data = "Datasets/AB_NYC_2019.csv"
lat_lng_newyork_data_df = pd.read_csv(
    lat_lng_newyork_data)
lat_lng_newyork_data_df.head()
```

```python
# Visualize
lat_lng_newyork_data_df = lat_lng_newyork_data_df.reset_index()
lat_lng_newyork_data_df.set_index('id', inplace=True)
del lat_lng_newyork_data_df['index']
lat_lng_newyork_data_df.head()
```

```python
"""
#using the cg library
census_block_list = []
for lat, lon in zip(lat_lng_newyork_data_df.loc[:, 'latitude'], lat_lng_newyork_data_df.loc[:, 'longitude']):
        #print(lat,lon)
        census_block_data = cg.coordinates(x=lon, y=lat)
        #print(census_block_data)
        data = census_block_data['2010 Census Blocks'][0]['GEOID']
        #print(data)
        #print("\n")
        census_block_list.append(data)
        time.sleep(1)

"""

base_url = "https://geo.fcc.gov/api/census/area?"
census_block_list = []
for lat, lon in zip(lat_lng_newyork_data_df.loc[:, 'latitude'], lat_lng_newyork_data_df.loc[:, 'longitude']):
        
    try:
        #print(lat,lon)
        census_block_url = f'{base_url}lat={lat}&lon={lon}&format=json'
        #print(census_block_url)
        response = requests.get(census_block_url)
        census_block_data = response.json()
        print(census_block_data['results'][0]['block_fips'])
        census_block_list.append(census_block_data['results'][0]['block_fips'])

    except Exception as e:
        print(e)
        time.sleep(3)
        continue
   
```

```python
lat_lng_newyork_data_df['census_block'] = census_block_list
lat_lng_newyork_data_df.head()
```
